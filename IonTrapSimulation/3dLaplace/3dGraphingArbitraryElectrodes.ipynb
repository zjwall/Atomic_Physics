```python
%matplotlib inline
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
n=50 ##defines fineness of mesh
plt.rcParams['figure.figsize'] = [12, 8]

###create a cylinder


R = np.linspace(0, 1, n)
u = np.linspace(0,  2*np.pi, n)

z = np.outer(R, np.cos(u))
y = np.outer(R, np.sin(u))
h1 = np.linspace(-10, -10, n)
h2 = np.linspace(10, 10, n)
h3= np.linspace(-10,10,n)
x1 = np.repeat([h1],n,axis=0)
x2 = np.repeat([h2],n,axis=0)
fdfdfs,x3=np.meshgrid(h3,h3)
zh = np.repeat(z[n-1:],int(n),axis=0)
yh = np.repeat(y[n-1:],int(n),axis=0)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x3,yh,zh, rstride=1, cstride=1,
                color='r', edgecolor='k')
ax.plot_surface(x1,y,z, rstride=1, cstride=1,
                color='r', edgecolor='k')
ax.plot_surface(x2,y,z, rstride=1, cstride=1,
                color='r', edgecolor='k')


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
ax.view_init(30, 0) ##sets the angle of view elevation, azimuthal
fig
```

```python
##copy cyclinder 4 times around axis 
n=50
d = 2
rad = 1
l=10
center = [[d,d],[-d,d],[d,-d],[-d,-d]]

fig = plt.figure()
ax = plt.axes(projection='3d')
for i in center:    
    R = np.linspace(0, rad , n)
    u = np.linspace(0,  2*np.pi, n)

    z = np.outer(R, np.cos(u))
    y = np.outer(R, np.sin(u))
    z=z-i[0]
    y=y-i[1]
    h1 = np.linspace(-l, -l, n)
    h2 = np.linspace(l, l, n)
    h3= np.linspace(-l,l,n)
    x1 = np.repeat([h1],n,axis=0)
    x2 = np.repeat([h2],n,axis=0)
    f,x3=np.meshgrid(h3,h3)
    zh = np.repeat(z[n-1:],int(n),axis=0)
    yh = np.repeat(y[n-1:],int(n),axis=0)


    ax.plot_surface(x3,yh,zh, rstride=1, cstride=1,
                color='r', edgecolor='k')
    ax.plot_surface(x1,y,z, rstride=1, cstride=1,
                color='r', edgecolor='k')
    ax.plot_surface(x2,y,z, rstride=1, cstride=1,
                color='r', edgecolor='k')
```

```python

###create a sphere

n=50 
R = 4
u = np.linspace(0,  2*np.pi, n*2)
v = np.linspace(0, np.pi, n*2)
phi, theta = np.meshgrid(u,v)

x = R*np.sin(phi)*np.cos(theta)
y = R*np.sin(phi)*np.sin(theta)
z = R*np.cos(phi)

# plt.plot(x, y, marker='.', color='k', linestyle='none')
# plt.show()



fig = plt.figure()
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(x,y,z, rstride=1, cstride=1,
                color='r')
ax.scatter(x,y,z, 
                color='b', s=.05)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.view_init(30, 0) ##sets the angle of view elevation, azimuthal
ax.plot_surface(x,y,z, rstride=1, cstride=1,
                color='r')
ax.scatter(x,y,z, 
                color='b', s=1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')



plt.show()
```
