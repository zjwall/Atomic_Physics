```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import fftpack
%matplotlib inline
file = open("out-20.txt","r")
x = list()
y = list()
z = list()
t = list()
for line in file:
    fields = line.split(" ")
    if fields[0] == '\n':
        continue
    tt = fields[0][4:]
    xt = fields[2][2:]
    yt = fields[4][2:]
    zt = fields[6][2:]
    if float(tt) > 30:
        t.append(float(tt))
        x.append(float(xt))
        y.append(float(yt))
        z.append(float(zt))
    
    

    
f_s = 1
rotated_y = list()
for l in range(len(y)):
    rotated_y.append(y[l]*np.cos(np.pi/3)-z[l]*np.sin(np.pi/3))

plt.plot(t,x, color ='indigo')
plt.grid()
plt.show()

plt.plot(t,rotated_y, color ='indigo')
plt.grid()

plt.show()
plt.plot(t,z, color ='indigo')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq
# plt.rcParams['figure.figsize'] = [12, 4]
# plt.rcParams.update({'font.size': 22})



# Number of sample points
N = len(t)
# sample spacing
T = 1.0e-3
temp = np.linspace(0.0, N*T, N, endpoint=False)
x = x-np.mean(x)
xf = fft(x)
tf = fftfreq(N, T)[:N//2]
fig = plt.figure()
fig.suptitle('Z-axis Position', fontsize=20)
ax = fig.gca()
ax.patch.set_facecolor('#ababab')
ax.patch.set_alpha(0.5)
plt.xlabel("time(us)")
plt.ylabel("position")
plt.plot(temp,x, color ='indigo')
plt.grid()
plt.show()

fig = plt.figure()
fig.suptitle('Z-axis Fourier Transform', fontsize=20)
ax = fig.gca()
ax.patch.set_facecolor('#ababab')
ax.patch.set_alpha(0.5)
plt.xlabel("frequency (MHz)", fontsize=15)
plt.ylabel("Weight")
plt.plot(tf[:200], 2.0/N * np.abs(xf[0:N//2])[:200], color ='indigo')
plt.grid()
plt.show()



# Number of sample points
N = len(t)
# sample spacing
T = 1.0e-3
temp = np.linspace(0.0, N*T, N, endpoint=False)
y = y-np.mean(y)
yf = fft(y)
tf = fftfreq(N, T)[:N//2]

fig = plt.figure()
fig.suptitle('Y-axis Position', fontsize=20)
ax = fig.gca()
ax.patch.set_facecolor('#ababab')
ax.patch.set_alpha(0.5)
plt.xlabel("time(us)")
plt.ylabel("position")
plt.plot(temp,y, color ='indigo')
plt.grid()
plt.show()

fig = plt.figure()
fig.suptitle('Y-axis Fourier Transform', fontsize=20)
ax = fig.gca()
ax.patch.set_facecolor('#ababab')
ax.patch.set_alpha(0.5)
plt.xlabel("frequency (MHz)", fontsize=15)
plt.ylabel("Weight")
plt.plot(tf[:200], 2.0/N * np.abs(yf[0:N//2])[:200], color ='indigo')
plt.grid()
plt.show()


# Number of sample points
N = len(t)
# sample spacing
T = 1.0e-3
temp = np.linspace(0.0, N*T, N, endpoint=False)
z = z-np.mean(z)
zf = fft(z)
tf = fftfreq(N, T)[:N//2]

fig = plt.figure()
fig.suptitle('X-axis Position', fontsize=20)
ax = fig.gca()
ax.patch.set_facecolor('#ababab')
ax.patch.set_alpha(0.5)
plt.xlabel("time(us)")
plt.ylabel("position")
plt.plot(temp,z, color ='indigo')
plt.grid()
plt.show()


fig = plt.figure()
fig.suptitle('X-axis Fourier Transform', fontsize=20)
ax = fig.gca()
ax.patch.set_facecolor('#ababab')
ax.patch.set_alpha(0.5)
plt.xlabel("frequency (MHz)", fontsize=15)
plt.ylabel("Weight")
plt.plot(tf[:200], 2.0/N * np.abs(zf[0:N//2])[:200], color ='indigo')
plt.grid()
plt.show()

```
