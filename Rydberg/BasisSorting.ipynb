```python
##Our goal is to divide the arrays out and sort them by their true basis quantum number 
from arc import *
num_points = 100
calc = StarkMap(Caesium())
calc.defineBasis(41, 1.0, 0.5, 0.5,50, 60, 6)
calc.diagonalise(np.linspace(2000,2500,num_points),progressOutput=True,upTo = 6, totalContributionMax = .99999999999)
print("%.5f MHz cm^2 / V^2 " % calc.getPolarizability())
```

```python
#lets first just graph all the s and p levels with multicolor to check the crossings

from scipy import constants as sc
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]



# Sweep the electric field and calculate energy levels (in GHz)
electric_fields = calc.eFieldList  # Electric fields applied in V/m
energy_levels = calc.y[0]  # Stark shifted energy levels in GHz
basis = calc.basisStates

new_y = np.zeros((len(electric_fields),len(energy_levels)))
for i in range(len(electric_fields)):
    for k in range(len(calc.y[0])):
#         if basis[i][1] == 0:
#             pass
#             #plt.plot(electric_fields/100, [j[i] for j in calc.y])
        new_y[i][calc.composition[i][k][0][1]] = calc.y[i][k]

for i in range(len(calc.y[0])):
    if basis[i][1] == 0:
        plt.scatter(electric_fields/100, [j[i] for j in new_y], color='b')
    if basis[i][1] == 1:
        plt.scatter(electric_fields/100, [j[i] for j in new_y], color='g')
    if basis[i][1] == 2:
        plt.scatter(electric_fields/100, [j[i] for j in new_y], color='r')
    if basis[i][1] == 3:
        plt.scatter(electric_fields/100, [j[i] for j in new_y], color='c')
    if basis[i][1] == 4:
        plt.scatter(electric_fields/100, [j[i] for j in new_y], color='m')
    if basis[i][1] == 5:
        plt.scatter(electric_fields/100, [j[i] for j in new_y], color='y')
    if basis[i][1] == 6:
        plt.scatter(electric_fields/100, [j[i] for j in new_y], color='k')  

            
plt.xlim(20,25)
plt.ylim(-1110,-1125)
plt.show()
```
