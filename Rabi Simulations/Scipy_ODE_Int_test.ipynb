```python
%matplotlib inline
import time
import numpy as np
ti=time.time()
def pend(y, t, b, c,wl):
    theta, omega = y
    dydt = [omega, omega*(np.cos(wl*t))-theta*np.cos(wl*t)**2]
    return dydt

W=.125*2*np.pi*10e6
wl=50
b = 0.25
c = 5.0
y0 = [1, 0.0]
t = np.linspace(0, 10, 100000)
t1 =time.time()

from scipy.integrate import odeint
sol = odeint(pend, y0, t, args=(b, c,wl))
t2= time.time()
import matplotlib.pyplot as plt

plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()



tf=time.time()
print('time', tf-ti)
print('time', t1-ti)
print('time', t2-t1)
print('time', tf-t2)
```
