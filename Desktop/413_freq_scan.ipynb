```python
x = [257400,257700,
257890,
257800,
258200,
258300,
258400,
258600,
258900]

y = [0,1,
1,
2,
4,
4,
3,
3,
0
]
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from scipy.special import factorial
plt.rcParams['figure.figsize'] = [12, 8]

def func(x, *params):
    y = np.zeros_like(x)
    ctr = params[0]
    amp = params[1]
    wid = params[2]
    y = y + amp * np.exp( -((x - ctr)/wid)**2)
    return y

guess = [x[4], 4, 200]

popt, pcov = curve_fit(func, x, y, p0=guess)
x_fit = np.linspace(x[0],x[-1],100000)
fit = func(x_fit, *popt)

print(popt)
##Plotting Results###
fig = plt.figure()
fig.suptitle('413 Freq Vs Ions Loaded', fontsize=20)
ax = fig.gca()
ax.ticklabel_format(useOffset=False)
ax.patch.set_facecolor('#ababab')
ax.patch.set_alpha(0.5)
plt.xlabel("413 Frequency")
plt.ylabel("Ions Loaded")
plt.plot(x,y,'o', color ='indigo')
plt.plot(x_fit,fit, color ='crimson')
plt.legend(loc=(1.04,.51))
plt.grid()
plt.show()
```
