```python
###Lorentzian data fitting


import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from scipy.special import factorial
from scipy.optimize import curve_fit
import numpy as np



data1 = np.loadtxt('Z:/Group_Share/Barium/Data/2022.dir/02.dir/2022_02_01.dir/00316 - E2LaserSweep_prob.csv', delimiter = ',')
data2 = np.loadtxt('Z:/Group_Share/Barium/Data/2022.dir/02.dir/2022_02_01.dir/00319 - E2LaserSweep_prob.csv', delimiter = ',')
data3 = np.loadtxt('Z:/Group_Share/Barium/Data/2022.dir/02.dir/2022_02_01.dir/00322 - E2LaserSweep_prob.csv', delimiter = ',')
data4 = np.loadtxt('Z:/Group_Share/Barium/Data/2022.dir/02.dir/2022_02_01.dir/00325 - E2LaserSweep_prob.csv', delimiter = ',')
data5 = np.loadtxt('Z:/Group_Share/Barium/Data/2022.dir/02.dir/2022_02_01.dir/00328 - E2LaserSweep_prob.csv', delimiter = ',')
data6 = np.loadtxt('Z:/Group_Share/Barium/Data/2022.dir/02.dir/2022_02_01.dir/00331 - E2LaserSweep_prob.csv', delimiter = ',')
data7 = np.loadtxt('Z:/Group_Share/Barium/Data/2022.dir/02.dir/2022_02_01.dir/00334 - E2LaserSweep_prob.csv', delimiter = ',')
data8 = np.loadtxt('Z:/Group_Share/Barium/Data/2022.dir/02.dir/2022_02_01.dir/00343 - E2LaserSweep_prob.csv', delimiter = ',')
data9 = np.loadtxt('Z:/Group_Share/Barium/Data/2022.dir/02.dir/2022_02_01.dir/00349 - E2LaserSweep_prob.csv', delimiter = ',')
data10 = np.loadtxt('Z:/Group_Share/Barium/Data/2022.dir/02.dir/2022_02_01.dir/00352 - E2LaserSweep_prob.csv', delimiter = ',')


freq = data10[:,0]
prob = 1-data10[:,1]

def func(x, *params):
    y = np.zeros_like(x)
    ctr = params[0]
    amp = params[1]
    wid = params[2]
    y = y + amp * np.exp( -((x - ctr)/wid)**2)
    return y

guess = [5649.985, .3, .003]
popt, pcov = curve_fit(func, freq, prob, p0=guess)
x_fit = np.linspace(freq[0],freq[-1],100000)
fit = func(x_fit, *popt)

##Plotting Results###
plt.rcParams['figure.figsize'] = [12, 8]
fig = plt.figure()
fig.suptitle(' Data', fontsize=20)
ax = fig.gca()
ax.ticklabel_format(useOffset=False)
ax.patch.set_facecolor('lightgrey')
ax.patch.set_alpha(.5)
plt.xlabel("EOM Frequency (MHz)")
plt.ylabel("$D_{5/2}$ Probability ")

plt.plot(freq,prob,'o', color ='indigo',label = 'Blue Data')
plt.plot(x_fit,fit, color ='crimson',label = 'Red Data')


plt.legend(loc=(1.04,.51))

plt.grid()
plt.show()
```
